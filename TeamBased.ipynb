```python
# Codificar times, lado
# Achar media
#df.set_index('gameid', inplace=True)

#Funo para criar a mdia das n partidas anteriores de determinada equipe
def adjust_average(database, interval, teamName):
    team_matches = database.groupby('team').get_group(teamName)
    index = team_matches.index
    
    for i in index[::-1]:
        idx = index.get_loc(i)
        if idx >= interval:
            avrg = team_matches.kills[idx-interval:idx].mean()
        else:
            avrg = team_matches.kills[:idx].mean()
        
        #database.at[i,['kills']] = float(avrg)
        
    return team_matches

#Funo para criar a mdia das n partidas anteriores de todas as equipes
def teams_average_interval(database, k):
    average_data = pd.DataFrame()
    teams = database.groupby('team')
    
    pos = 0
    for name, team in teams:
        interval = team[-k:]
        if(len(interval == k)):
            avg = interval.mean()
            avg['team'] = name

            average_data = average_data.append(avg, ignore_index=True)  

    average_data = average_data.dropna(axis=1)
    
    return average_data

#Funco para criar a mdia total das partidas de determinada equipe
def full_average(database, teamName):    
    team_matches = database.groupby('team').get_group(teamName)
    return team_matches.mean()

#Funo para criar uma tabela com todas as mdias das equipes do Mundial
def creates_team_avarages(database, teamName):
    teams_av = pd.DataFrame(columns=database.columns)
    
    for team in teamName:
        aux = full_average(database, team)
        aux['team'] = team
        aux['game'] = np.nan
        aux['result'] = np.nan
        aux['playoffs'] = np.nan
        teams_av = teams_av.append(aux, ignore_index=True)

    return teams_av

#Funo para criar os jogos do mundial
def create_game(blue_team, red_team, result, game, playoff):
        blue = teams_avg[teams_avg['team'] == blue_team]
        red = teams_avg[teams_avg['team'] == red_team]
                        
        blue['game'] = game
        blue['result'] = result
        blue['playoffs'] = playoff                
        red['game'] = game
        red['playoffs'] = playoff

        common_columns = ['game', 'playoffs']
        
        match = pd.merge(blue, red, left_on=common_columns, right_on=common_columns)
        match.drop(columns_to_drop, axis=1, inplace=True)
                        
        return match
        
```
