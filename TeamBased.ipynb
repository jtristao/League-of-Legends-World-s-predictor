```python
#df.set_index('gameid', inplace=True)

#Função para criar a média das n partidas anteriores de determinada equipe (não terminada)
def adjust_average(database, interval):
    teams = database['team'].unique()

    average_table = pd.DataFrame()

    for team in teams:
        team_matches = database.groupby('team').get_group(team)
        team_average = team_matches[-interval::].mean()
        team_average = team_average.to_frame().T
        team_average['team'] = team

        average_table = pd.concat([average_table, team_average], axis=0)
    
    # Change columns order so team name is at the begginig
    columns = list(average_table.columns)
    columns.insert(2, 'team')
    columns = columns[:-1]
   
    average_table = average_table[columns]
    
    return average_table

#Funcão para criar a média total das partidas de determinada equipe
def full_average(database, teamName):    
    team_matches = database.groupby('team').get_group(teamName)
    return team_matches.mean()

#Função para produzir features resultantes da diferença de duas outras features
def feat_resultante(data_solo, data_duplo, columns):
    par = pd.DataFrame(columns=data_duplo.columns)
    impar = pd.DataFrame(columns=data_duplo.columns)
    par = data_duplo.iloc[::2]
    par.set_index('gameid', inplace=True)
    impar = data_duplo.iloc[1::2]
    impar.set_index('gameid', inplace=True)

    data_solo.set_index('gameid', inplace=True)
    diff = par[columns].sub(impar[columns])
    data_solo[columns] = diff[columns]
    
    return data_solo
        

#Função para criar uma tabela com todas as médias das equipes do Mundial
def creates_team_avarages(database, teamName):
    teams_av = pd.DataFrame(columns=database.columns)
    
    for team in teamName:
        aux = full_average(database, team)
        aux['team'] = team
        aux['game'] = np.nan
        aux['result'] = np.nan
        aux['playoffs'] = np.nan
        teams_av = teams_av.append(aux, ignore_index=True)

    return teams_av

#Função para criar os jogos do mundial
def create_game(blue_team, red_team, result, game, playoff):
        blue = teams_avg[teams_avg['team'] == blue_team]
        red = teams_avg[teams_avg['team'] == red_team]
                        
        blue['game'] = game
        blue['result'] = result
        blue['playoffs'] = playoff                
        red['game'] = game
        red['playoffs'] = playoff

        common_columns = ['game', 'playoffs']
        
        match = pd.merge(blue, red, left_on=common_columns, right_on=common_columns)
        match.drop(columns_to_drop, axis=1, inplace=True)
                        
        return match
    
#Função para criar os jogos do mundial com a diferença de features
def create_game_diff(blue_team, red_team, result, game, playoff):
        blue = teams_avg[teams_avg['team'] == blue_team]
        red = teams_avg[teams_avg['team'] == red_team]
                        
        blue['game'] = game
        blue['result'] = result
        blue['playoffs'] = playoff                
        red['result'] = 0
        red['game'] = 0
        red['playoffs'] = 0
        
        blue.set_index('gameid', inplace=True)
        red.set_index('gameid', inplace=True)

        #Dataframe com colunas que vão ser subtraídas
        clm_to_sub = list(blue.columns)
        clm_to_sub.remove('team')

        diff = blue[clm_to_sub].sub(red[clm_to_sub])
        blue[clm_to_sub] = diff[clm_to_sub]

        #Adicionando team names para facilitar futuras vizualizações
        blue['team_y'] = red['team']
        blue.rename(columns={'team':'team_x'}, inplace=True)
                        
        return blue
    
def avg_drags(x):
    if x['dragons (type unknown)'] > 4:
        x['elementaldrakes'] = 1
    else:
        x['elementaldrakes'] = 0

    if x['dragons (type unknown)'] > 5:
        x['elders'] = x['dragons (type unknown)'] - 4
    else:
        x['elders'] = 0

    drgs = x['dragons (type unknown)'] - x['elementaldrakes'] - x['elders']
    x['infernals', 'mountains', 'clouds', 'oceans'] = drgs/4
    
    return x
        
```
