```python
import pandas as pd
import seaborn as sn
import matplotlib.pyplot as plt
#pd.set_option("display.max_rows", None, "display.max_columns", None)
```

```python
# Dataset geral
match = pd.read_csv("Database/2020_LoL_esports_match_data_from_OraclesElixir_20201002.csv")
match.head()
```

```python
#filtra os times do mundial no ultimo split
times= ["Top Esports", "G2 Esports", "DAMWON Gaming", "Team SoloMid", "JD Gaming", "Suning", "Fnatic", "DRX", "Rogue", "Gen.G", "FlyQuest",
        "Machi Esports", "LGD Gaming", "Team Liquid", "PSG Talon", "Unicorns Of Love.CIS"]
leagues = ["LPL", "LEC", "LCS", "LCK", "PCS", "LCL"]

filtro = match[match['team'].isin(times)]
filtro = filtro[filtro['league'].isin(leagues)]
filtro = filtro[filtro['split']=='Summer']
```

```python
#seleciona todas as prtidas que esses times jogaram
id = pd.unique(filtro['gameid'])

df = match[match['gameid'].isin(id)]
df = df[df['player'].isna()]
df = df[df['league'].isin(leagues)]
df.head()
```
